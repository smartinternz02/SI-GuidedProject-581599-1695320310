```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\Users\nihar\Jupyter notebooks\WA_Fn-UseC_-HR-Employee-Attrition.csv')
df
```

```python
df.isnull().sum()
```

```python
cols=["BusinessTravel" ,"Department" ,"EducationField" ,"Gender" ,"JobRole" ,"MaritalStatus", "Over18" ,"OverTime","Attrition"]
newdf=pd.get_dummies(df[cols])
newdf
```

```python
df.drop(cols,axis=1,inplace=True)
df=pd.concat([df,newdf],axis=1)
df.drop("EmployeeNumber",inplace=True,axis=1)
df
```

```python
encodedcols=["HourlyRate", "MonthlyIncome", "MonthlyRate"]
encodeddf=df[encodedcols]
encodeddf
df
```

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
x_scaled=pd.DataFrame(ms.fit_transform(encodeddf),columns=encodeddf.columns)
x_scaled
```

```python
df.drop(encodedcols,axis=1,inplace=True)
df=pd.concat([df,x_scaled],axis=1)
df
```

```python
y=df["DailyRate"]
df.drop("DailyRate",axis=1,inplace=True)
y
```

```python
X=df
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state=16)
logreg.fit(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=50)
rf_model.fit(X_train, y_train)
```

```python
lrpred=logreg.predict(X_test)
clfpred=clf.predict(X_test)
rfpred=rf_model.predict(X_test)
```

```python
preddata= pd.DataFrame({'Y_test': y_test, 'Logistic Regression': lrpred, 'Decision Tree Classifier':clfpred, 'Random Forest': rfpred})
preddata
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, accuracy_score
```

```python
mse = mean_squared_error(y_test, lrpred)
rmse = mean_squared_error(y_test, lrpred, squared=False)
mae = mean_absolute_error(y_test, lrpred)
r2 = r2_score(y_test, lrpred)
print("Logistic Regression")
print("MSE: ", mse)
print("RMSE: ", rmse)
print("MAE: ", mae)
print("R2: ", r2)
```

```python
mse = mean_squared_error(y_test, clfpred)
rmse = mean_squared_error(y_test, clfpred, squared=False)
mae = mean_absolute_error(y_test, clfpred)
r2 = r2_score(y_test, clfpred)
print("Decision Tree Classifier")
print("MSE: ", mse)
print("RMSE: ", rmse)
print("MAE: ", mae)
print("R2: ", r2)
```

```python
mse = mean_squared_error(y_test,  rfpred)
rmse = mean_squared_error(y_test,  rfpred, squared=False)
mae = mean_absolute_error(y_test,  rfpred)
r2 = r2_score(y_test,  rfpred)
print("Random Forest")
print("MSE: ", mse)
print("RMSE: ", rmse)
print("MAE: ", mae)
print("R2: ", r2)
```
